This is for extract features from dataset

This notebook transformed the original data into new features and exported as csv file

```python
from functions import *
```

```python
# import original data
data1 = arff.loadarff(r'insects/InsectSound_TRAIN.arff')
train= pd.DataFrame(data1[0])

data2 = arff.loadarff(r'insects/InsectSound_TEST.arff')
test= pd.DataFrame(data2[0])

# get rid of empty data points
train = train[~train.duplicated(subset=train.columns[:-1], keep=False)]
test = test[~test.duplicated(subset=test.columns[:-1], keep=False)]

# for random seed for consistency
random_seed = 1222
```

```python
#combined_normalized Training and Test Set
combined = pd.concat([train, test], axis=0)
combined
```

```python
# split the data and labels
labels = combined.iloc[:,600:]
labels.reset_index(drop=True, inplace=True)
data = combined.iloc[:,:-1] 
data.reset_index(drop=True, inplace=True)
```

Baseline

```python
baseline_data = transform_stats(data)
baseline = pd.concat([baseline_data, labels], axis=1, ignore_index=True)
baseline.rename(columns={baseline.columns[-1]: 'target'}, inplace=True)
# export it as csv file
baseline.to_csv('features/baseline.csv', index=False)
```

Fast Fourier Transform

```python
fft_data= data.apply(calculate_fft, axis=1, result_type='expand')
fft = pd.concat([fft_data, labels], axis=1, ignore_index=True)
fft.rename(columns={fft.columns[-1]: 'target'}, inplace=True)
# export it as csv file
fft.to_csv('features/fft.csv', index=False)
```

Temporal Integration

```python
ti_data= data.apply(temp_int, axis=1, result_type='expand')
ti = pd.concat([ti_data, labels], axis=1, ignore_index=True)
ti.rename(columns={ti.columns[-1]: 'target'}, inplace=True)
# export it as csv file
ti.to_csv('features/ti.csv', index=False)
```

Spectral Rolloff

```python
roll_data= data.apply(spec_rollof, axis=1, result_type='expand')
roll = pd.concat([roll_data, labels], axis=1, ignore_index=True)
roll.rename(columns={roll.columns[-1]: 'target'}, inplace=True)
# export it as csv file
roll.to_csv('features/rolloff.csv', index=False)
```

Spectral Density Extraction

```python
psd_data= data.apply(psd, axis=1, result_type='expand')
psd = pd.concat([psd_data, labels], axis=1, ignore_index=True)
psd.rename(columns={psd.columns[-1]: 'target'}, inplace=True)
# export it as csv file
psd.to_csv('features/psd.csv', index=False)
```

Wavelets transform

```python
wt_data= data.apply(wt_transform, axis=1, result_type='expand')
wt = pd.concat([wt_data, labels], axis=1, ignore_index=True)
wt.rename(columns={wt.columns[-1]: 'target'}, inplace=True)
# export it as csv file
wt.to_csv('features/wavelets.csv', index=False)
```
